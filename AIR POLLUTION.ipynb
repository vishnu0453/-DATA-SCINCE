```python
import numpy as np
import pandas as pd
from  pandas import read_csv
data=read_csv("AirQualityUCI.csv", sep=';')
df=pd.DataFrame(data)
df
```

```python
# Q1 REPLACING COMMA WITH DOT 
# Q27 Q28 Q29 Q30

for i in range(df.columns.size):         #we are iterating over each and every column
    temp=list(str(df[df.columns[i]][0])) #df[df.columns[i]][0] means 0 th element in ith column of df, we are checking whether it consists comma or not
    x=','                                # for that we are typecasting that particular element to string and then converted into list
    if(x in temp):                      #if comma is present , we are replacing comas with dots for that particular column and storing back into that column
        df[df.columns[i]]=df[df.columns[i]].str.replace(',','.')
df
# df.columns.size
```

```python
# Q2 CONVERTING DATA IN COLUMNS TO FLOAT
df['T']=df['T'].astype(float)
df['RH']=df['RH'].astype(float)
df['AH']=df['AH'].astype(float)
df['CO(GT)']=df['CO(GT)'].astype(float)
df['C6H6(GT)']=df['C6H6(GT)'].astype(float)
df

```

```python
#Q32
cols=[]
for i in range(2,15):
    cols.append(df.columns[i]) #appending all columns in cols list
no_garbage_vals=[]
for i in range(2,15):
    x=df[df[df.columns[i]]==-200]     #appending number of rows with  -200 in a particular column into no_garbage_vals list
    no_garbage_vals.append(x.shape[0])
no_garbage_vals
garbage_df=pd.DataFrame(columns=['Column Name', 'No of rows with garbage values(-200)'])
# created a df for cols and no_garbage_vals
garbage_df[garbage_df.columns[0]]=cols
garbage_df[garbage_df.columns[1]]=no_garbage_vals
garbage_df
```

```python
#Q33
garbage_percent=[]
for i in range(garbage_df.shape[0]):
    x=garbage_df[garbage_df.columns[1]][i]
    result=(x/df.shape[0])*100   #dividing no 0f rows i a column with garb vlas by total no fo rows in df(df.shape[0])
#                                   and multiplying it with 100 to find percentage of garb vals in that column
    garbage_percent.append(result)
garbage_percent
percentage_of_garbage_vals_in_the_column=pd.DataFrame(garbage_percent)
percentage_of_garbage_vals_in_the_column
garbage_df['Percentage of garbage vals in the column']=percentage_of_garbage_vals_in_the_column
garbage_df  #creating new column in garbage_df to that consists of percentage of garb columns

```
